```python
def get_solution(sol,smr):
    plane = np.zeros((6,10))
#     print(sol)
    for s in sol:
        shape = smr[s[0]-1][s[1]-1]
        for s2 in shape:
            plane[s[2]+s2[0]-1,s[3]+s2[1]-1] = s[0]-1
    return plane
        
```
